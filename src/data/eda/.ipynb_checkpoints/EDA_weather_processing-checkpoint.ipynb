# EDA Code for Weather Processing
This code is primarily to inform how to truncate the weather data, and to generate some visualizations for the PowerPoint/Medium article.

Import all libraries used.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from datetime import datetime, timedelta
from functools import reduce
from scipy.spatial.distance import cdist
from math import radians, sin, cos, sqrt, atan2
import seaborn as sns
import folium
from PIL import Image
import io
```

Set plot style for <code>matplotlib</code>.

```python
plt.style.use('fivethirtyeight')
```

Let's get a list of all available .csv files first.

```python
csv_list = ['../weather/download/' + filename for filename in os.listdir('../weather/download') if filename.endswith('.csv')]
csv_list
```

After that, let's get a list of all weather station names.

```python
station_names = set()

for filename in csv_list:
    df = pd.read_csv(filename)
    station_names =station_names.union(set(df['Station Name'].unique()))

station_names = list(station_names)
print(station_names)
```

Let's create a list of dataframes, one corresponding to each weather station.

```python
weather_stations = {}

rename = {
        'Station Name': 'STATION_NAME',
        'Longitude (x)': 'x',
        'Latitude (y)': 'y',
        'Precip. Amount (mm)': 'PRECIP_AMOUNT',
        'Temp (°C)': 'TEMP',
        'Dew Point Temp (°C)': 'DEW_POINT_TEMP',
        'Rel Hum (%)': 'RELATIVE_HUMIDITY',
        'Stn Press (kPa)': 'STATION_PRESSURE',
        'Wind Dir (10s deg)': 'WIND_DIRECTION',
        'Wind Spd (km/h)': 'WIND_SPEED',
        'Date/Time (UTC)': 'UTC_DATE'
    }

for filename in csv_list:
    df = pd.read_csv(filename, parse_dates=['Date/Time (UTC)']).rename(columns=rename)[rename.values()]

    station_name = df['STATION_NAME'][0]

    if not df.empty:
        if station_name in weather_stations:
            weather_stations[station_name] = pd.concat([weather_stations[station_name], df], axis=0)
        else:
            weather_stations[station_name] = df
                    
start_date = datetime(2012, 1, 1)
cutoff_date = datetime(2023, 11, 1)

for station_name, df in weather_stations.items():
    # Filter DF so that UTC date is after the start date and before the cutoff date.
    df = df[(df['UTC_DATE'] >= start_date) & (df['UTC_DATE'] < cutoff_date)]
    
    df = df.sort_values(by=['UTC_DATE'])
    
    # REINDEX
    # Create a DateTimeIndex with hourly frequency for the desired range
    new_index = pd.date_range(start=start_date, end=(cutoff_date - timedelta(hours=1)), freq='1H')

    # Reindex the DataFrame with the new DateTimeIndex and keep missing values as NaN for 'value' column
    new_df = df.set_index('UTC_DATE').reindex(new_index).reset_index()
    new_df = new_df.rename(columns={'index': 'UTC_DATE'})
    
    weather_stations[station_name] = new_df
    print(f'Station: {station_name}\tNew Length: {len(new_df)}\tOld Length: {len(df)}')

    
weather_stations = list(weather_stations.values())
```

Let's consider data imputation by __Inverse Distance Weighing (IDW)__. However, there needs to be enough observations available for it to be effective. Let's plot the number of stations with some variable missing for each hour.

Let's plot everything at once to visualize., considering the maximum missing for ANY feature for a particular hour.

```python
plt.figure(figsize=(8, 6), dpi=300)

missing = []

for feature in ['PRECIP_AMOUNT',
       'TEMP', 'DEW_POINT_TEMP', 'RELATIVE_HUMIDITY',
       'STATION_PRESSURE', 'WIND_DIRECTION', 'WIND_SPEED']:
    
    missing_feature = []
    
    for df in weather_stations:
        missing_feature.append(df[feature].isna().to_numpy())
    
    missing_feature = np.column_stack(missing_feature)
    missing_feature = missing_feature.sum(axis=1)
    missing.append(missing_feature)

missing = np.column_stack(missing)
missing = missing.max(axis=1)
    
temp_df = pd.DataFrame({'UTC_DATE': weather_stations[0]['UTC_DATE'], 'MISSING': missing})
temp_df = temp_df.groupby(pd.Grouper(key='UTC_DATE', freq='1M'))['MISSING'].max()
    
plt.step(temp_df.index, temp_df, lw=2)

plt.xlabel('Date')
plt.ylabel('Max # of stations missing')
plt.title('Missing data value count')
plt.tight_layout()
plt.savefig('../../../images/weather_missing_data.png')
```

This shows that after <code>2014-11-05 19:00:00</code>, the data should be of sufficient quality to work with through IDW imputation. 

Next, after that datetime, let's check the completeness of each column for each station. Essentially, if a variable like TEMP is __majority NAs__, it just suggests that was unobserved at the weather station, and we should not even consider a TEMP input from that example weather station.

```python
def get_missing(feature):
    missing_feature = []
    
    for df in weather_stations:
        missing_feature.append(df[feature].isna().to_numpy())
    
    missing_feature = np.column_stack(missing_feature)
    return missing_feature.sum(axis=1)

index_first = np.where(get_missing('TEMP') > 9)[0][-1]
date_first = weather_stations[0].iloc[index_first]['UTC_DATE']

print('First index:', index_first, 'First date:', date_first)

trimmed_weather_stations = [df.iloc[(index_first + 1):] for df in weather_stations]
```

```python
for df in trimmed_weather_stations:
    
    station_name = df['STATION_NAME'].iloc[df['STATION_NAME'].first_valid_index()]
    df = df.reset_index(drop=True)
    
    print(station_name)
    print(df[['PRECIP_AMOUNT', 'TEMP', 'DEW_POINT_TEMP', 'RELATIVE_HUMIDITY', 'STATION_PRESSURE', 'WIND_DIRECTION', 'WIND_SPEED']].isna().mean())
    print(len(df))
    print()
```

This shows that PRECIP_AMOUNT is ignored in ST CATHARINES A and WIND_DIRECTION and WIND_SPEED are ignored in TORONTO CITY. 

Let us make a basic visualization for the PowerPoint presentation.

```python
trunc_weather_stations = [df.iloc[30000:30048] for df in trimmed_weather_stations]
```

```python
plt.figure(figsize=(10, 6), dpi=600)

for station_name, df in zip(station_names, trunc_weather_stations):
    plt.plot(df['UTC_DATE'], df['TEMP'], color='b', alpha=0.2)
ax = plt.gca()   
ax.axis('off')  # Hide axes
ax.set_xticks([])  # Hide x-axis ticks
ax.set_yticks([])  # Hide y-axis ticks
ax.set_frame_on(False)

plt.xlabel('Date')
plt.ylabel('Temperature')
plt.savefig('../../../images/temperature_lines.png')
```

Let's create another visualization for the PowerPoint.

```python
# Creating 5 subplots in a stacked manner with no vertical space
fig, axs = plt.subplots(5, 1, figsize=(5, 5), sharex=True)
df = weather_stations[0].iloc[50000:70000]

# Plotting on each subplot
axs[0].plot(df['UTC_DATE'], df['PRECIP_AMOUNT'], lw=0.4, color='white')
axs[1].plot(df['UTC_DATE'], df['RELATIVE_HUMIDITY'], lw=0.4, color='white')
axs[2].plot(df['UTC_DATE'], df['TEMP'], lw=0.4, color='white')
axs[3].plot(df['UTC_DATE'], df['DEW_POINT_TEMP'], lw=0.4, color='white')
axs[4].plot(df['UTC_DATE'], df['WIND_SPEED'], lw=0.4, color='white')

# Adding labels, title, legend to each subplot
fig.patch.set_facecolor('none')
for ax in axs:
    ax.set_facecolor('none')
    
    ax.set_ylabel('Y-axis', color='white')
    ax.xaxis.label.set_color('white')
    ax.yaxis.set_visible(False)  # Hide y-axis labels
    ax.xaxis.set_visible(False)  # Hide y-axis labels
    ax.tick_params(axis='x', colors='white')
    ax.tick_params(axis='y', colors='white')
    ax.grid()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

axs[-1].set_xlabel('Time', color='white')
    
# Adjusting layout to remove vertical space between subplots
plt.tight_layout(pad=0.0)

# Show the plot
plt.savefig('../../../images/weather_data.svg', transparent=True)
```

Let's create a map visualizing all the locations of the stations.

```python
coordinates = []
station_names = []

for df in weather_stations:
    x_series = df['x']
    y_series = df['y']
    name_series = df['STATION_NAME']
    
    coordinates.append([y_series.iloc[y_series.first_valid_index()], x_series.iloc[x_series.first_valid_index()]])
    station_names.append(name_series.iloc[name_series.first_valid_index()])

coordinates
```

Save the map twice, once with markers, and once without.

```python
m = folium.Map(location=[44, -79.39611111], zoom_start=8)

# Save the folium map as a png file.
img_data = m._to_png(5)
img = Image.open(io.BytesIO(img_data))
img.save('../../../images/map_no_markers.png')

# Display the map
m
```

```python
m = folium.Map(location=[44, -79.39611111], zoom_start=8)

# Add markers for each latitude and longitude pair to the map
for coord in coordinates:
    folium.Marker(location=coord, icon=folium.Icon(color='red')).add_to(m)

# Save the folium map as a png file.
img_data = m._to_png(5)
img = Image.open(io.BytesIO(img_data))
img.save('../../../images/map_markers.png')

# Display the map
m
```
