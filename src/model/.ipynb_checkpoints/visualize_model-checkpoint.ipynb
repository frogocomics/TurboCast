# Visualize Model Outputs

Import all libraries used.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Set plot style for <code>matplotlib</code>.

```python
plt.style.use('fivethirtyeight')
visual_directory = '../../images'
```

There are six visualizations which are interesting, and illustrate the performance and limitations of the model.
1. Learning curve plot
2. Predictions vs. ground truth
3. CNN intermediate layer visualizations
4. Model loss comparisons
5. Loss vs. training horizon
6. Train set precipitation distribution
7. Precipitation over time

## Data fetching

```python
# Step 1
train_loss = np.load('outputs/train_losses.npy') * 100 ** 2  # Precipitation is scaled
                                                             # down by 100 originally, so need
                                                             # to scale the loss back up!
val_loss = np.load('outputs/val_losses.npy') * 100 ** 2

# Step 2
feature = np.load('outputs/test_inputs.npy')[:, :, -11]
target = np.load('outputs/test_targets.npy')
prediction = np.load('outputs/test_outputs.npy')

# Step 3
radar_image = np.load('outputs/image_input.npy')[0, 0, :, :]
cnn_processed1 = np.load('outputs/cnn_output1.npy')[0]
cnn_processed2 = np.load('outputs/cnn_output2.npy')[0]

# Step 4
# Values are from results from benchmark_performance.ipynb
avg_loss, persist_loss, rf_loss, xgboost_loss, dnn_loss = 0.32411015720409897, \
    0.20348833640270092, 0.15493580540369475, 0.11346626318842216, 0.1925132401083829

# Step 5
average_loss = np.load('outputs/length_losses.npy')
horizon = range(2, 16)
```

Quickly check shapes to make sure they are correct!

```python
print(feature.shape)
print(target.shape)
print(prediction.shape)
```

## 1. Learning curve

```python
n_epochs = len(train_loss)

plt.figure(figsize=(8, 6), dpi=300)

# Plot losses for train and val
plt.plot(range(1, 1 + n_epochs), train_loss, label='Train loss')
plt.plot(range(1, 1 + n_epochs), val_loss, label='Val loss')

# Plot labels
plt.title('Loss over time')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.ylim([0, 2])
plt.legend(loc=1)
plt.tight_layout()

plt.savefig(visual_directory + '\\learning_curve.svg')
plt.savefig(visual_directory + '\\learning_curve.png')
```

## 2. Prediction vs. Ground Truth

```python
def plot_vs_truth(feature, target, index, prediction):
    """
    Function to plot a model prediction against the ground truth.
    """
    
    feature, target, prediction = feature[index] * 100, target[index] * 100, prediction[index] * 100
    print(feature)
    print(target)
    print(prediction)
    feature_len = len(feature)
    target_len = len(target)
    
    plt.figure(figsize=(8, 3), dpi=300)
    
    plt.plot(feature, label='Ground truth')
    
    plt.plot(range(feature_len - 1, feature_len + target_len), np.insert(target, 0, feature[-1]), label='Target')
    
    plt.plot(range(feature_len - 1, feature_len + target_len), np.insert(prediction, 0, feature[-1]),
             label='Prediction')
    
    plt.legend(loc=1)
    plt.xlabel('Hour')
    plt.ylabel('Precipitation (mm/hr)')
    
    plt.tight_layout()

num = 79
plot_vs_truth(feature, target, num, prediction)

plt.savefig(visual_directory + f'\\pred{num}.png')
```

```python
num = 979
plot_vs_truth(feature, target, num, prediction)

plt.savefig(visual_directory + f'\\pred{num}.png')
```

## 3. CNN Intermediate Layer Visualizations

```python
fig = plt.figure(figsize=(8, 4), dpi=300)
# Add grid specifications
gs = fig.add_gridspec(2, 4)

axs = []

axs.append(fig.add_subplot(gs[:,0:2]))
axs.append(fig.add_subplot(gs[0,2]))
axs.append(fig.add_subplot(gs[0,3]))
axs.append(fig.add_subplot(gs[1,2]))
axs.append(fig.add_subplot(gs[1,3]))

# Plot the large image on the left
axs[0].imshow(radar_image)  # Show the large image on the first subplot (top-left)
axs[0].set_title('Input')
axs[0].axis('off')  # Hide axis for the large image

# Plot the small images on the right
for i, ax in enumerate(axs[1:]):
    ax.imshow(cnn_processed2[i])  # Show each small image
    ax.set_title(f'{i + 1}')
    ax.axis('off')  # Hide axis for small images

plt.tight_layout()

plt.savefig(visual_directory + '\\cnn_intermediate2.svg')
plt.savefig(visual_directory + '\\cnn_intermediate2.png')
```

```python
# Do something similar, but for the 1st CNN layer instead

fig = plt.figure(figsize=(8, 4), dpi=300)
# Add grid specifications
gs = fig.add_gridspec(2, 6)

axs = []

axs.append(fig.add_subplot(gs[:,0:2]))
axs.append(fig.add_subplot(gs[0,2]))
axs.append(fig.add_subplot(gs[0,3]))
axs.append(fig.add_subplot(gs[0,4]))
axs.append(fig.add_subplot(gs[0,5]))
axs.append(fig.add_subplot(gs[1,2]))
axs.append(fig.add_subplot(gs[1,3]))
axs.append(fig.add_subplot(gs[1,4]))
axs.append(fig.add_subplot(gs[1,5]))

# Plot the large image on the left
axs[0].imshow(radar_image)  # Show the large image on the first subplot (top-left)
axs[0].set_title('Input')
axs[0].axis('off')  # Hide axis for the large image

# Plot the small images on the right
for i, ax in enumerate(axs[1:]):
    ax.imshow(cnn_processed1[i])  # Show each small image
    ax.set_title(f'{i + 1}')
    ax.axis('off')  # Hide axis for small images

plt.tight_layout()

plt.savefig(visual_directory + '\\cnn_intermediate1.svg')
plt.savefig(visual_directory + '\\cnn_intermediate1.png')
```

## 4. Model loss comparison

```python
plt.figure(figsize=(8, 6), dpi=300)

# Plot losses for each model
plt.bar(['Average', 'Persistence', 'Random Forest', 'XGBoost', 'DNN'], [avg_loss, persist_loss, rf_loss, xgboost_loss, dnn_loss], width=0.6)

# Plot labels
plt.xlabel('Method')
plt.ylabel('Mean test loss')
plt.tight_layout()

plt.savefig(visual_directory + '\\method_comparison.png')
```

## 5. Loss vs. input horizon

```python
plt.figure(figsize=(8, 4), dpi=300)

# Scaled up loss
scaled = average_loss * 100 ** 2

plt.plot(horizon, scaled.astype(float))

# Plot labels
plt.title('Loss vs. number of input steps')
plt.xlabel('Input steps')
plt.ylabel('Mean loss')

plt.ylim([0.18, 0.22])
plt.tight_layout()
plt.savefig(visual_directory + '\\loss_vs_input.png')
```

## 6. Precipitation distribution

```python
plt.figure(figsize=(8, 6), dpi=300)

# Load and plot the entire precipitation distribution (combined train/val/test)
df = pd.read_csv('../data/weather/processed/TORONTO CITY.csv')
plt.hist(df['PRECIP_AMOUNT'], bins=100)

# Plot labels
plt.yscale('log')
plt.xlabel('Precipitation (mm/hr)')
plt.ylabel('Count')
plt.title('Precipitation distribution')
plt.tight_layout()

plt.savefig(visual_directory + '\\precip_distribution.png')
```

Notice just how skewed the distribution of precipitation is. It is so skewed, we need the y-axis in log scale just to be able to see the distribution clearly!

## 7. Precipitation over time

```python
plt.figure(figsize=(8, 3), dpi=300)

# Plot precipitation for test set only
df = pd.read_csv('../data/weather/processed/TORONTO CITY.csv', parse_dates=['UTC_DATE'])
plt.plot(df[df['UTC_DATE'].dt.year == 2023]['UTC_DATE'], df[df['UTC_DATE'].dt.year == 2023]['PRECIP_AMOUNT'], lw=1)

# Plot labels
plt.ylabel('Precipitation (mm/hr)')
plt.title('Test set precipitation')

plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator())
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))

plt.tight_layout()

plt.savefig(visual_directory + '\\precip_over_time.png')
```
